```python
# Проверяем содержимое столбцов
columns = ["year", "price_in_euro", "power_kw", "power_ps"]

for col in columns:
    # приводим к строке и убираем пробелы
    values = df[col].astype(str).str.strip()

    # оставляем только значения, которые НЕ являются числами
    invalid = values[
        (~values.str.fullmatch(r"\d+")) &
        (values != "nan") &
        (values != "None")
    ]

    print(f"\n{col}")
    print(f"Некорректных значений: {len(invalid)}")

    if len(invalid):
        print(invalid.unique()[:20])   # первые 20 различных значений
```

Удалить все строки, где числовые поля имеют неверный формат

```python
import pandas as pd

mask = (
    df["year"].astype(str).str.fullmatch(r"\d{4}") &
    df["price_in_euro"].astype(str).str.fullmatch(r"\d+") &
    (
        df["power_kw"].isna() |
        df["power_kw"].astype(str).str.fullmatch(r"\d+")
    ) &
    (
        df["power_ps"].isna() |
        df["power_ps"].astype(str).str.fullmatch(r"\d+")
    )
)

print(f"Удаляется строк: {(~mask).sum()}")

df = df.loc[mask].copy()
```

Преобразовать столбцы в числовой тип

```python
numeric_cols = [
    "year",
    "price_in_euro",
    "power_kw",
    "power_ps"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

2. data auditing

```python
df = pd.read_csv("raw/data.csv")
```

```python
def column_report(df):
    report = {}

    for col in df.columns:
        s = df[col].astype(str).str.strip()

        report[col] = {
            "unique_values_sample": s.value_counts().head(10),
            "nan_count": df[col].isna().sum(),
        }

    return report

report = column_report(df)
```

```python
import pandas as pd

def find_non_numeric(df, col):
    s = df[col].astype(str).str.strip()

    # оставляем только строки, которые НЕ являются числом
    bad = s[~s.str.match(r"^-?\d+([.,]\d+)?$")]

    return bad.value_counts()
```

```python
numeric_candidates = [
    "year",
    "price_in_euro",
    "power_kw",
    "power_ps",
    "fuel_consumption_l_100km",
    "fuel_consumption_g_km",
    "mileage_in_km"
]

for col in numeric_candidates:
    print("\n", col)
    print(find_non_numeric(df, col).head(20))
```

```python
def detect_shifted_rows(df):
    mask = pd.Series(False, index=df.index)

    for col in df.columns:
        s = df[col].astype(str)

        # если в числовой колонке встречается буква → подозрительно
        mask |= s.str.contains(r"[A-Za-zÄÖÜäöü]", na=False)

    return df[mask]

bad_rows = detect_shifted_rows(df)

print(len(bad_rows))
```

```python
checks = {
    "year": r"^\d{4}$",
    "price_in_euro": r"^\d+$",
    "power_kw": r"^\d+(\.\d+)?$",
    "power_ps": r"^\d+(\.\d+)?$",
    "mileage_in_km": r"^\d+(\.\d+)?$"
}

for col, pattern in checks.items():
    bad = df[~df[col].astype(str).str.match(pattern, na=False)]
    print(col, "bad rows:", len(bad))
```
